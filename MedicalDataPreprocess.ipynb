```python
import numpy as np
from sklearn.datasets import load_breast_cancer
import csv

data = load_breast_cancer()
d = data['data']
print(d.shape)
print(data['feature_names'])

list(data.target_names)



pd = np.zeros_like(d, int)

for i in range(d.shape[1]):
  min = np.min(d[:, i])
  max = np.max(d[:, i]) - min
  mean = np.mean(d[:, i])
  pd[:, i] = d[:, i] >= mean
  #pd[:, i] = np.ceil(100*(d[:, i]-min)/max)
print(pd, np.max(pd, 0), np.min(pd, 0))

print(pd.shape, data['target'].shape)

nt = np.concatenate((data['target'].reshape(-1, 1), np.zeros((140, 1))), 0)

#q = np.concatenate((pd, nt), 1)

q = np.concatenate((pd, data['target'].reshape(-1, 1)), 1)


print(q.shape)

# Write the document vectors to csv files
def writeToCsv(outPath, corpus):
    f = open(outPath, "a", newline='')
    w = csv.writer(f, dialect='excel')
    for x in corpus:
        w.writerow(x)
    f.close()

np.random.seed(20)
np.random.shuffle(q)

train = q[:470, :]
test = q[470:, :]

writeToCsv("train1.csv", train)
writeToCsv("test1.csv", test)


np.random.seed(21)
np.random.shuffle(q)

train = q[:470, :]
test = q[470:, :]

writeToCsv("train2.csv", train)
writeToCsv("test2.csv", test)

np.random.seed(22)
np.random.shuffle(q)

train = q[:470, :]
test = q[470:, :]

writeToCsv("train3.csv", train)
writeToCsv("test3.csv", test)
```
